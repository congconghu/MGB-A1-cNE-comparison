### get frequency span of recordings and cNEs

```python
su_sig = su[su.strf_sig]
freq_span = su_sig.groupby(by=['exp', 'probe'])['bf'].agg(min_freq = 'min', max_freq = 'max')
freq_span['freq_span_oct'] = np.log2(freq_span.max_freq / freq_span.min_freq)
sessions = sessions.join(freq_span, on=['exp', 'probe'])
sessions.to_json(session_file)
```

```python
freq_span.head()
```

```python
ne['member1'] = ne['pattern1'].apply(lambda x: np.where(x > 1/np.sqrt(len(x)))[0])
ne['member2'] = ne['pattern2'].apply(lambda x: np.where(x > 1/np.sqrt(len(x)))[0])
ne['member_all'] = ne[['member1', 'member2']].apply(np.concatenate, axis=1).apply(np.unique)
ne['member_share'] = ne[['member1', 'member2']].apply(lambda x: np.intersect1d(*x), axis=1)
freq_span_all = []
freq_span_share = []
for i in range(len(ne)):
    pair = ne.iloc[i]
    exp = pair.exp
    probe = pair.probe
    members = su_sig[(su_sig.exp == exp) & (su_sig.probe == probe)]
    # get frequency span of combined members
    members_all = members[members['index'].apply(lambda x: x in pair.member_all)]
    if len(members_all) > 1:
        freq_span_all.append(np.log2(members_all.bf.max() / members_all.bf.min()))
    else:
        freq_span_all.append(np.nan)
    # get frequency span of shared members
    members_share = members[members['index'].apply(lambda x: x in pair.member_share)]
    if len(members_share) > 1:
        freq_span_share.append(np.log2(members_share.bf.max() / members_share.bf.min()))
    else:
        freq_span_share.append(np.nan)
ne['freq_span_oct_all'] = freq_span_all
ne['freq_span_oct_share'] = freq_span_share
ne.to_json(ne_file)
```
